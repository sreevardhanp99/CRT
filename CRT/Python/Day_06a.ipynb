**How to access private members outside of a class?**

We can access private members from outside of a class using the following two approaches.

1)public method to acess private member

2)Name mangling to access private members

Syntax:

print(objName._ClassName__privateVariableName)

```python
class Rectangle:
  __length=0 #private variable
  __breadth=0
  def __init__(self):
    #constructor
    self.__length=5
    self.__breadth=3
    print(self.__length)
    print(self.__breadth)
rect = Rectangle()
print("\n")
print(rect._Rectangle__length) # ERROR
print(rect._Rectangle__breadth)
```

```python
class MyClass:
  def __init__(self,private_member):
    self.__private_member =  private_member

  def get_private_number(self):
    return self.__private_member

  def set_private_member(self,new_value):
    self.__private_member= new_value


obj = MyClass(42)

value = obj.get_private_number()
print(value)

obj.set_private_member(100)
new_value = obj.get_private_number()
print(new_value)
```

* Abstraction is a technique to hide the implementation details and show only the necessary features of an object.
* Pre-defined functions are similar to data abstraction


**ABC (Abstract Base Class)**
* The ABC module is part of the python standard library and is used for creating abstract base classes.
* An abstract base class is a class that cannot be instantiated on its own and is meant to be subclassed by other classes.
* It often defines abstract methods that must be implemented by its subclasses.

```python
from abc import ABC,abstractmethod
class shape(ABC):
  def area(self):
    pass

class circle(shape):
  def __init__(self,radus):
    self.radus=radus

  def area(self):
    return 3.14*self.radus*self.radus

class square(shape):
  def __init__(self,side):
    self.side=side

  def area(self):
    return self.side**2

c=circle(2)
c.area()
s=square(5)
s.area()
```

**Question 4**

Given an integer array nums and an integer k, returns the number of pairs(i,j) where i < j such that | nums[i] - nums[j] | == k
The value of |x| is defined as:

x if x >= 0

-x if x<= 0

INPUT: nums = [1,2,2,1] , k = 1

OUTPUT: 4

Explanation: The pairs with an absolute difference of 1

```python
#04 - a
class Solution:
  def abs_diff_pairs(self,num,target):
    c=0
    for i in range(0,len(num)):
      for j in range(0,len(num)):
        if i < j:
          if abs(num[i]-num[j]) == target:
            c+=1
    return c
num = list(map(int,input().split(",")))
target = int(input())
obj = Solution()
print(obj.abs_diff_pairs(num,target))
```

```python
class Solution: #05
  def target_index(self,num,target):
    if target not in num:
      num.append(target)
      num.sort()
    return num.index(target)
num = list(map(int,input().split(",")))
num.sort()
print(num)
target = int(input())
obj = Solution()
print(obj.target_index(num,target))
```
