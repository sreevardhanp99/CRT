```python
def binaryConversion(n):  # 01
  str1 = ""
  print("Number = ",n)
  while(n>0):
    r = n % 2
    str1 += str(r)
    n //= 2
  int_str = int(str1[::-1])
  print("Binary number : ")
  return int_str
num=int(input("Enter the number : "))
binaryConversion(num)
```

```python
def fibonacciSeries(number):  # 04 - sir code
  num1, num2 = 0, 1
  result = []
  print("Fibonacci Series ")
  for i in range(number):
    result.append(num1)
    num3 = num1 + num2
    num1 = num2
    num2 = num3
  return result
num = int( input())
r = fibonacciSeries(num)
for val in r:
  print(val, end= " ")
```

## OOPS
* OOPS in programming stand for object oriented programming system.
* It is a programming methodology to design a program using classes and objects
* OOPS treats every entity as an object.
* we can create as many objects as we want for a given class.
### **class**
* A class just defines the structure of how anything should look.
* It does not point to anything or anyone in particular.
* For example, say, HUMAN is a class, which has suppose --name, age, gender, city.
* It does not point to any specific HUMAN out there. but yes, it explains the properties and functuions any HUMAN should or any object of class HUMAN should have.

### **How to define a class in python?**
* classes in python can be defined by the keyword class, which is followed by the name of class and a colon.
* example :
class Human:
 // statements



### **What is an __init__ method?**
* The properties that all class objects must have should be defined in a method called init()
* The _ _init_ _ method in python is a special method that is called when an object is created from a class.
* Every time a new class object is created, _ _init_ _() sets the initial state of the object by assigning the values we provide inside the objects properites.
* That is, _ _ init_ _() initializes each new instance of the class.


### **self**
* self is not a keyword. Self is object name.
1. _ _init_ _() can take any number of parameters, but the first parameter is always a variable called **self**.
2. The **self** parameter is a reference to the current instance of the class.
3. It means, the self parameter points to the address of the current object of a class, allowing us to access the data of its(the object's) variables.
4. We can use any name in place of **self**, but it has to be the fist parameter of any function in the class.


### **Attributes**
* There are 2 types of attributes in python:
1. class attributes
2. instance attributes


### **Creating an Object in class**
* When we create a new object from a class, it is called instantiating an object.
* An object can be instantiated by the class name followed by the parentheses.
* We can assign the object of a class to any variable.


**syntax:**

objectName = className()
