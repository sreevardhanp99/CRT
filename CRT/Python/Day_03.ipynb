```python
def primeNo(num):
  flag=0
  for i in range(2,int(num ** 0.5)):
    if (num%i==0):
        flag=1
        break

  if(flag==0 and num > 1):
      print('{0} is a prime number'.format(num))
  else:
    print(f'{num} is not a prime')
n=int(input())

primeNo(n)

```
