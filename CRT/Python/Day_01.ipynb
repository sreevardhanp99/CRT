```python
class Solution:
  def armStrog(self,num):
    num_str = str(num)
    count = len(num_str)
    s = 0
    for digit in num_str:
      s += int(digit)**count
    if s == num :
      print("Yes")
    else:
      print("No")
num = int(input())
obj = Solution()
print(obj.armStrog(num))
```

```python
class stack(object):
    def __init__(self):
        self.stack=[]
        self.index=0
    def isEmpty(self):
        return self.stack==[]
    def push(self,data):
        self.stack.insert(self.index,data)
        self.index+=1
        return f"{data} is inserted into stack "
    def pop(self):
        self.index-=1
        data=self.stack.pop(self.index)
        return f'{data} is deleted from stack'
    def stackSize(self):
        return len(self.stack)
if __name__=="__main__":
    s=stack()
    print(s.isEmpty())
    print(s.push(10))
    print(s.push("hello"))
    print(s.push(30))
    print(s.push(40))
    print(s.stackSize())
    print(s.isEmpty())
    print(s.pop())
    print(s.pop())
    print(s.pop())
```
